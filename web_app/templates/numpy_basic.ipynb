```python
import pandas as pd

s = pd.Series([10, 20, 30], index=['a','b','c'])
print(s['b'])  # Access by label
data = {'Name':['Alice','Bob','Charlie'],
        'Age':[25,30,22]}
df = pd.DataFrame(data)
print(df)
import numpy as np

df.loc[1, 'Age'] = np.nan  # Introduce missing value

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Drop missing values
df.dropna(inplace=True)
print(df)
```

```python
data = {'Class':['A','A','B','B','A'],
        'Score':[85,90,70,75,88]}
df = pd.DataFrame(data)

grouped = df.groupby('Class')['Score'].mean()
print(grouped)
pivot = df.pivot_table(index='Class', values='Score', aggfunc='mean')
print(pivot)
df.sort_values(by='Score', ascending=False, inplace=True)
df['Rank'] = df['Score'].rank(ascending=False)
print(df)
# Apply a function to a column
df['Score_Squared'] = df['Score'].apply(lambda x: x**2)
print(df)
```
